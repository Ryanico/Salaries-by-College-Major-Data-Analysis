```python
#How would we calculate the difference between the earnings of the 10th and 90th percentile?
#Well, Pandas allows us to do simple arithmetic with entire columns, so all we need to do is use the .subtract()
```

```python
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
#The output of this computation will be another Pandas dataframe column. We can add this to our existing dataframe with the .insert() method.
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

```python
#To see which degrees have the smallest spread, we can use the .sort_values() method. 
#And since we are interested in only seeing the name of the degree and the major, we can pass a list of these two column names to look at the 
#.head() of these two columns exclusively.
```

```python
low_risk = clean_df.sort_values('Spread')
```

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#Using the .sort_values() method, can you find the degrees with the highest potential? 
#Find the top 5 degrees with the highest values in the 90th percentile.
```

```python
highest_potential = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
```

```python
highest_potential[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

```python
#Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation?
```

```python
greatest_spread = clean_df.sort_values('Spread', ascending=False)
```

```python
greatest_spread[["Undergraduate Major", "Spread"]].head()
```

```python
# Oftentimes you will want to sum rows that belong to a particular category. For example, which category of degrees has the highest average salary? 
#Is it STEM, Business, or HASS (Humanities, Arts, and Social Science)? 

#To answer this question, we need to learn to use the .groupby() method. This allows us to manipulate data similar to a Microsoft Excel Pivot Table.
#We have three categories in the 'Group' column: STEM, HASS and Business. Let's count how many majors we have in each category.
```

```python
clean_df.groupby('Group').count()
```

```python
#Now, can you use the .mean() method to find the average salary by group?
```

```python
clean_df.groupby('Group').mean(numeric_only=True)
```
